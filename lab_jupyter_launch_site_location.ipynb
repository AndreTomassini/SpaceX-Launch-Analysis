The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
# 1) Load the CSV (works in both JupyterLite and local environments)
import io, pandas as pd
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv"

try:
    from js import fetch                          # JupyterLite
    resp = await fetch(URL)
    b = await resp.arrayBuffer()
    df = pd.read_csv(io.BytesIO(bytearray(b)))
except Exception:
    # Ambiente local/Colab
    try:
        df = pd.read_csv(URL)
    except Exception:
        import urllib.request
        urllib.request.urlretrieve(URL, "spacex_launch_geo.csv")
        df = pd.read_csv("spacex_launch_geo.csv")

# 2) Check column names
print(df.columns.tolist())

# 3) Normalize column names and create the expected subset
cols = [c.lower().strip() for c in df.columns]
df.columns = cols

# Accept either 'lat/long' or 'latitude/longitude'
if "lat" in df.columns and "long" in df.columns:
    lat_col, lon_col = "lat", "long"
elif "latitude" in df.columns and "longitude" in df.columns:
    lat_col, lon_col = "latitude", "longitude"
else:
    raise ValueError("Latitude/Longitude columns not found in the CSV.")

# 'class' usually already exists; if it comes as 'Class' in another dataset, standardize:
# df.rename(columns={"Class": "class"}, inplace=True)

spacex_df = df[["launch site", lat_col, lon_col, "class"]].copy()
spacex_df.rename(columns={lat_col: "Lat", lon_col: "Long", "launch site": "Launch Site"}, inplace=True)

spacex_df.head()
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# --- Coastline point (use the corrected lat/lon you captured on the waterline) ---
# Example values you measured with MousePosition:
coastline_lat = 28.56847   # <-- put your final coastline latitude here
coastline_lon = -80.56902  # <-- put your final coastline longitude here

# Launch site coordinates (already defined earlier in the notebook)
# launch_site_lat = ...
# launch_site_lon = ...

# Recalculate great-circle distance to the coastline (in km)
distance_coastline = round(
    calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon), 2
)

# --- Coastline marker with distance label ---
marker_coast = folium.Marker(
    location=(coastline_lat, coastline_lon),
    icon=folium.DivIcon(
        html=f"""
        <div style="font-size: 12px; color: teal;">
            <b>{distance_coastline:.2f} km</b> to Coastline
        </div>
        """
    )
)
site_map.add_child(marker_coast)

# --- PolyLine from launch site to coastline (keep style minimal like the lab) ---
line_coast = folium.PolyLine(
    locations=[(launch_site_lat, launch_site_lon), (coastline_lat, coastline_lon)],
    color='teal',
    weight=2
)
site_map.add_child(line_coast)

# (Optional) tiny endpoints to highlight targets, matching your style
for (lat, lon, color) in [
    (coastline_lat, coastline_lon, 'teal'),
]:
    folium.CircleMarker((lat, lon), radius=5, color=color, fill=True, fill_opacity=0.9).add_to(site_map)

# Display map and numeric output (so the notebook shows the corrected value)
from IPython.display import display
display(site_map)
print({"Coastline (km)": distance_coastline})
```
